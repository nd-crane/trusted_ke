```python
import pandas as pd
import ast
```

```python
with open("test_nyt_2k.jsonl.hyps") as f:
    lines = f.readlines()
lines = [line.strip() for line in lines]
```

```python
data = pd.read_csv('../../OMIn_dataset/data/FAA_data/Maintenance_Text_data_nona.csv')
```

```python
out_dict = {'c5_id':[],'doc_idx':[], 'c119_input':[],'head':[],'relation':[],'tail':[]}

for doc_idx, line in enumerate(lines):
    if line != '':
        for triple in line.split(') ('):
            head, relation, tail = [part.strip() for part in triple.replace(')','').replace('(','').split(';')]
            out_dict['c5_id'].append(data['c5'].iat[doc_idx])
            out_dict['doc_idx'].append(doc_idx)
            out_dict['c119_input'].append(data['c119'].iat[doc_idx])
            out_dict['head'].append(head)
            out_dict['relation'].append(relation)
            out_dict['tail'].append(tail)
```

```python
out_df = pd.DataFrame(out_dict)
out_df
```

```python
out_df.to_csv('../../tool_results/deepstruct/deepstruct_nyt.csv',index=False)
```
