```python
def get_annotator_ents(name):
    gold_standard_path = 'ner.csv'

    # load output from the tool to be evaluated
    data = pd.read_csv(gold_standard_path)
    
    
    # rename columns
    columns={"c5_unique_id": "id", "c119_text": "sample", name:"entities"}
    data.rename(columns=columns, inplace=True)
    
    data = data[['id', 'sample', 'entities']]
    
    # Remove spaces within entities, split by comma, then explode
    data['entities'] = data['entities'].str.replace(', ', ',').str.split(',')
    data = data.explode('entities').dropna().reset_index(drop=True)

    return data
```

```python
alyssa = get_annotator_ents('Alyssa')
jonathan = get_annotator_ents('Jonathan')
kate = get_annotator_ents('Kate')
```
